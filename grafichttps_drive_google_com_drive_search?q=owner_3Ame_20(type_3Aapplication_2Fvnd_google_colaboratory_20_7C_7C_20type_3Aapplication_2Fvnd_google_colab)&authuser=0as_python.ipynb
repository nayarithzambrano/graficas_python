##GRAFICAS EN DOS DIMENCIONES

### Grafica de una lista


```python
# Importamos la libreria
# import matplotlib.pyplot

from matplotlib.pyplot import plot, show ,xlim ,ylim,xticks
```

```python
# Grafica No.1

y = [1,2,3,4,5]

plot(y)
show()
```

```python
# Grafica No.2
y = [1,2,3,4,5]

plot(y, "*")
show()
```

```python
# Grafica No.3

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plot(x, y, "*")
show()
```

```python
# Grafica No.4

# Graficas con limites

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plot(x, y, "*")
xlim(-10,10)
ylim(0 , 15)
show()
```

```python
# Grafica No.5

from numpy import linspace, pi ,sin

x =  linspace(-10*pi, 10*pi, 2560)
y = sin(x)

plot(x, y)
ylim(-2,2)
show()
```

```python
# Grafica No.6

from numpy import linspace, pi ,sin

x =  linspace(-pi,pi, 256)
y = sin(x)

plot(x, y)
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```

```python
# Grafica No.6
from numpy import linspace, pi ,sin, cos
from matplotlib.pyplot import plot, show ,xlim ,ylim,xticks, legend

x =  linspace(-pi,pi, 256)
y = sin(x)
z = cos(x)
plot(x, y, label = "seno")
plot(x, z, label = "coseno")
legend(loc="upper left")
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```

```python
# Grafica No.6
from numpy import linspace, pi ,sin, cos
from matplotlib.pyplot import plot, show ,xlim ,ylim,xticks, legend, figure

x =  linspace(-pi,pi, 256)
y = sin(x)
z = cos(x)

#Figura 1
a = figure(2)
plot(x, y, label = "seno")
legend(loc="upper left")
#Figura 20
b = figure(20)
plot(x, z, label = "coseno")
legend(loc="upper left")
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```

### EJEMPLOS GRAFICAS TRIDIMENCIONALES


```python

#IMPORTAMOS MDULOS NECESARIOS.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

#CREAMOS FIGURA.
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

#COORDENADA DE CADA BARRA
x = [1,2,3,4,2,6,7,8,9,10]
y = [3,4,6,7,1,2,5,9,10,5]
z = np.zeros(10)

dx = np.ones(10) #ANCHURA DE CADA BARRA
dy = np.ones(10) #PROFUNDIDAD DE CADA BARRA
dz = [2,3,4,3,4,5,3,7,8,2] #ALTURA DE CADA BARRA

#REPRESENTACION DE LOS DATOS
ax1.bar3d(x, y, z, dx, dy, dz)

plt.show()
```
